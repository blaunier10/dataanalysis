#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

```python

boston_df = pd.DataFrame(boston_df)

# 1. Boxplot para a variável "MEDV"
plt.figure(figsize=(10, 6))
plt.boxplot(boston_df['MEDV'])
plt.title('Boxplot for MEDV (Median value of owner-occupied homes)')
plt.ylabel('MEDV ($1000)')
plt.grid(True)
plt.show()

# 2. Gráfico de barras para a variável "CHAS"
plt.figure(figsize=(10, 6))
boston_df['CHAS'].value_counts().plot(kind='bar')
plt.title('Bar chart for CHAS (Charles River)')
plt.xlabel('CHAS (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 3. Boxplot para "MEDV" vs "AGE"
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=[0, 35, 70, 100], labels=['<= 35', '35-70', '> 70'])

plt.figure(figsize=(10, 6))
boston_df.boxplot(column='MEDV', by='AGE_group')
plt.title('Boxplot for MEDV vs AGE')
plt.suptitle('')
plt.xlabel('AGE Group')
plt.ylabel('MEDV ($1000)')
plt.grid(True)
plt.show()

# 4. Gráfico de dispersão para "NOX" vs "INDUS"
plt.figure(figsize=(10, 6))
plt.scatter(boston_df['NOX'], boston_df['INDUS'])
plt.title('Scatter plot for NOX vs INDUS')
plt.xlabel('NOX (parts per 10 million)')
plt.ylabel('INDUS (proportion of non-retail business acres per town)')
plt.grid(True)
plt.show()

# 5. Histograma para a variável "PTRATIO"
plt.figure(figsize=(10, 6))
plt.hist(boston_df['PTRATIO'], bins=10, edgecolor='black')
plt.title('Histogram for PTRATIO (pupil-teacher ratio by town)')
plt.xlabel('PTRATIO')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
